```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("medical_examination.csv")
```

```python
df
```

```python
bmi=df["weight"]/((df["height"]/100)**2)
```

```python
bmi
```

```python
df["Overweight"]=bmi.apply(lambda x:0 if x<25 else 1 )
```

```python
df["BMI"]=bmi
```

```python
df
```

```python
df["cholesterol"]=df["cholesterol"].apply(lambda x:0 if x==1 else 1)
df["gluc"]=df["gluc"].apply(lambda x:0 if x==1 else 1)
```

```python
df
```

```python
df_cat=pd.melt(df,id_vars="cardio",var_name="variable" , value_vars=['cholesterol','gluc', 'smoke', 'alco', 'active','Overweight'])
```

```python
df_cat
```

```python
df_cat=pd.melt(df,var_name="variable" , value_vars=['cholesterol','gluc', 'smoke', 'alco', 'active','Overweight'],id_vars="cardio")
```

```python
figure=sns.catplot(data=df_cat,kind="count",x="variable",hue="value",col="cardio").set_axis_labels("variables","total")
```

```python
df_heat=df[(df["ap_lo"]<=df["ap_hi"]) & (df["height"]>=df["height"].quantile(0.025)) & (df["height"]<=df["height"].quantile(0.975)) & (df["weight"]>=df["weight"].quantile(0.025)) & (df["weight"]<=df["weight"].quantile(0.975)) ]    
```

```python
df_heat
```

```python
correlation_matrix=df_heat.corr()
correlation_matrix
```

```python
#masking the upper triangle
mask=np.triu(correlation_matrix)
mask
```

```python
fig, ax= plt.subplots(figsize=(20,20))
sns.heatmap(correlation_matrix,mask=mask,fmt=".1f",vmax=.3,linewidths=0,square=True,cbar_kws={'shrink':0.5},annot=True,center=0)
```
